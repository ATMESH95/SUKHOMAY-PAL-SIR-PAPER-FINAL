```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('training_normalized.csv')
```

```python
df2=pd.read_csv('testing_normalized.csv')
```

```python
df3=pd.read_csv('testing.csv')
```

```python
#df1.head()
```

```python
X_train=df1.drop(['Experiment no.','UTS (Mpa)'],axis=1)
```

```python
y_train=df1['UTS (Mpa)']
```

```python
X_test=df2.drop(['Experiment no.','UTS (Mpa)'],axis=1)
```

```python
y_test=df2['UTS (Mpa)']
```

```python
y_test_actual=df3['UTS (Mpa)']
```

```python
X_train.shape
```

```python
model=Sequential()
model.add(Dense(8,input_dim=8))
model.add(Dense(18,input_dim=8,activation='relu'))
model.add(Dense(20,input_dim=18,activation='relu'))
model.add(Dense(1,activation='linear'))
weights = model.get_weights()
```

```python
keras.optimizers.Adam(learning_rate=0.00006,beta_1=0.9,beta_2=0.999,amsgrad=False)
```

```python
model.compile(loss='mean_squared_error',optimizer='adam',metrics=['mse'])
```

```python
model.summary()
```

```python
history=model.fit(X_train,y_train,epochs=30000,verbose=1)
```

```python
prediction=model.predict(X_test)
```
