```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('training.csv')
```

```python
df2 = pd.read_csv('testing.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar_X = MinMaxScaler()
scalar_Y = MinMaxScaler()
```

```python
X_train = df.drop(['Experiment no.','UTS (Mpa)'],axis=1)
```

```python
y_train = df[['UTS (Mpa)']]
```

```python
X_train  = scalar_X.fit_transform(X_train)
y_train = scalar_Y.fit_transform(y_train)
```

```python
X_test = df2.drop(['Experiment no.','UTS (Mpa)'],axis=1)
y_test = df2[['UTS (Mpa)']]
```

```python
X_test = scalar_X.transform(X_test)
y_test = scalar_Y.transform(y_test)
```

```python
X_train.shape
```

```python
inputs = keras.Input(shape = (8))
x = layers.Dense(8,activation='relu',kernel_regularizer=keras.regularizers.l2(0.01),bias_initializer=keras.initializers.random_normal())(inputs)
x = layers.Dense(13,activation='relu',kernel_regularizer=keras.regularizers.l2(0.01),bias_initializer=keras.initializers.random_normal())(x)
x = layers.Dense(18,activation='relu',kernel_regularizer=keras.regularizers.l2(0.01),bias_initializer=keras.initializers.random_normal())(x)
output = layers.Dense(1)(x)
model = keras.Model(inputs,output)
print(model.summary())
```

```python
model.compile(loss=keras.losses.MeanSquaredError(),optimizer = keras.optimizers.RMSprop(learning_rate=0.006,rho=0.9),metrics=['mse'])
```

```python
history = model.fit(X_train,y_train,epochs=300,verbose=2,validation_data=(X_test,y_test))
#model.evaluate(X_test,y_test)
```

```python
pre = model.predict(X_test)
```
